```python
# clase Automovil, atributo, marca, placa, precio
# funciones leerdatos(), mostrardatos()
# crear 3 objetos carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca: ")
        self.placa = input("Ingrese la placa: ")
        self.precio = float(input("Ingrese el precio: "))

    def mostrardatos(self):
        print("Marca:", self.marca)
        print("Placa:", self.placa)
        print("Precio:", self.precio)

# fin de la clase automovil
# creando los objetos carro1, carro2, carro3

print("carro nimero 1")
carro1 = Automovil(None, None, None)  # creando carro umero 1
carro1.leerdatos()
carro1.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("carro nimero 2")
carro2 = Automovil(None, None, None) # creando carro numero 2
carro2.leerdatos()
carro2.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("carro nimero 3")
carro3 = Automovil(None, None, None) # creando carro numero 3
carro3.leerdatos()
carro3.mostrardatos ()
print(" software realiizado por Davi Barrera NL = 8")
```
